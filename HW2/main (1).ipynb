```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
datamatches=pd.read_json("C:/Users/User/Desktop/ADM/HW2/Files/matches_England.json")
datalabels=datamatches['label'] #this is the column for matches and scores
datalabels 
```

```python
datalabels.isnull().sum() #we don't have any missing values which is good
```

```python
a=[]
for i in datalabels:
    i = i.replace('-','$$$').replace(',','$$$').split('$$$')
    a.append(i)
a
#splitted the data from '-' and ',' and appended them into an empty list
```

```python
for i in a:
    i[0]=i[0][:-1]
    i[1]=i[1][+1:]
    i[2]=int(i[2])
    i[3]=int(i[3])
a#fixed the blanks on each element of the list
```

```python
x=[]
for i in a:
    x.append(i[0])
datamatches['team1']=x
y=[]
for i in a:
    y.append(i[1])
datamatches['team2']=y
z=[]
for i in a:
    z.append(i[2])
datamatches['team1goal']=z
t=[]
for i in a:
    t.append(i[3])
datamatches['team2goal']=t
datamatches.drop('label',axis=1)#dropped the columnd and 
#finally appended the teams and scored individually to the main dataframe
```

```python
weekno=max(datamatches['gameweek'])
teamslist=list(datamatches['team1'].unique())
scores={}
for team in teamslist:
    scores[team]=(weekno)*[0]
#created empty dictionary for each team for 38 matches so that we can manipulate later
for week in range(weekno):#this is a loop for manipulating every team's every element for 38 weeks
    week_data = datamatches[datamatches['gameweek']==(week+1)]
    for i in range(week_data.shape[0]):
        s1 = week_data['team1goal'].to_list()
        s2 = week_data['team2goal'].to_list()
        t1=week_data['team1'].to_list()[i]
        t2=week_data['team2'].to_list()[i]
        if s1[i]>s2[i]:#if first team scored better than the second
            scores[t1][week]+=3#add three points to first
        elif s1[i]<s2[i]:#if second team scored better than the first
            scores[t2][week]+=3#add three points to second
        else:#if it's a tie
            scores[t1][week]+=1#add one point to each
            scores[t2][week]+=1  
```

```python
scores
```

```python
cumulative_scores={}
key=scores.keys()
lst=[]
for team in key:
    cumulative_scores[team]=np.cumsum(scores[team])#cumulative sum for each team
import collections
value=cumulative_scores.values()
freqlist=[]
for team in key:
         temp=collections.Counter(cumulative_scores[team])#occurence of each point in cumulative sum
         freqlist.append(temp)
for i in range(len(freqlist)):#printing the maximum occurence for each team's points
    print(teamslist[i],max(freqlist[i].values()))  
#the lowest are west bromwich albion and crystal palace
#highest ones are manchester united and manchester city(also validated from plot)
```

```python
gameweeks=list(datamatches['gameweek'].unique())
plt.figure(figsize=(10,10))
for team in cumulative_scores:
    plt.plot(cumulative_scores[team],label=team,linewidth=2)
plt.xticks(range(0,len(gameweeks)))#fixing x axis' ticks
plt.legend()
plt.grid()
```

```python
#a function for fuutball pitch
def draw_pitch(ax):
    # focus on only half of the pitch
    #Pitch Outline & Centre Line
    Pitch = mpatches.Rectangle([0,0], width = 120, height = 80, fill = False)
    #Left, Right Penalty Area and midline
    LeftPenalty = mpatches.Rectangle([0,22.3], width = 14.6, height = 35.3, fill = False)
    RightPenalty = mpatches.Rectangle([105.4,22.3], width = 14.6, height = 35.3, fill = False)
    midline = mpatches.ConnectionPatch([60,0], [60,80], "data", "data")

    #Left, Right 6-yard Box
    LeftSixYard = mpatches.Rectangle([0,32], width = 4.9, height = 16, fill = False)
    RightSixYard = mpatches.Rectangle([115.1,32], width = 4.9, height = 16, fill = False)


    #Prepare Circles
    centreCircle = plt.Circle((60,40),8.1,color="black", fill = False)
    centreSpot = plt.Circle((60,40),0.71,color="black")
    #Penalty spots and Arcs around penalty boxes
    leftPenSpot = plt.Circle((9.7,40),0.71,color="black")
    rightPenSpot = plt.Circle((110.3,40),0.71,color="black")
    leftArc = mpatches.Arc((9.7,40),height=16.2,width=16.2,angle=0,theta1=310,theta2=50,color="black")
    rightArc = mpatches.Arc((110.3,40),height=16.2,width=16.2,angle=0,theta1=130,theta2=230,color="black")
    
    element = [Pitch, LeftPenalty, RightPenalty, midline, LeftSixYard, RightSixYard, centreCircle, 
               centreSpot, rightPenSpot, leftPenSpot, leftArc, rightArc]
    for i in element:
        ax.add_patch(i)
fig=plt.figure() #set up the figures
fig.set_size_inches(7, 5)
ax=fig.add_subplot(1,1,1)
draw_pitch(ax) #overlay our different objects on the pitch
plt.ylim(-2, 90)
plt.xlim(-2, 130)
plt.axis('off')

sns.kdeplot(Ronaldo['xfirst'],Ronaldo["yfirst"], shade=True,n_levels=50) #highlighting the plot with Ronaldo's movements
plt.ylim(-2, 82)
plt.xlim(-2, 122)
plt.show()
```
